```python
#extract date from time_stamp
#https://stackoverflow.com/questions/33834727/how-do-you-extract-only-the-date-from-a-python-datetime
from datetime import datetime

engage['time_stamp'] = [datetime.date() for datetime in engage["time_stamp"]]
```

```python
print(len(engage.visited))
```

```python
engage = engage.sort_values('user_id')
engage.head()
```

```python
engage['time_stamp'] = pd.to_datetime(engage.time_stamp)
engage.dtypes
```

```python
engage = engage.set_index('time_stamp')
engage
```

```python
#create a rolling window over 7 days and look at the visited column
engage = engage.groupby('user_id').rolling('7D').sum()
```

```python
engage
```

```python
#create adopt df from visitors with at least 3 visits per 7 day period
adopt = engage[engage['visited'] >= 3]
adopt = adopt.drop('user_id', axis=1)
adopt = adopt.reset_index()
adopt
```

```python
#create array of unique users, change this into a list
adopted_user=adopt.user_id.unique()
adopted_user=adopted_user.tolist()
#adopted_user
```

```python
#preparing engage df for the new column
engage = engage.drop('user_id', axis=1)
engage = engage.reset_index()
engage
```

```python
#make adopted list and append to it either 0 or 1 for every user_id
adopted = []
for i in range(len(engage.user_id)):
    if engage['user_id'][i] in adopted_user:
        adopted.append(1)
    else:
        adopted.append(0)
#print(adopted)
```

```python
#create new column adopted in engage
engage['adopted'] = adopted
engage
```

```python
users = users.rename(columns={'object_id':'user_id'})
users.head()
```

```python
#left join onto users table
df_left = pd.merge(users, engage, on='user_id', how='left')
df_left
```

```python
#separate into two dataframes to make comparisons over adoption
non_adopted = df_left[df_left['adopted']==0]
non_adopted.head()
```

```python
adopted = df_left[df_left['adopted']==1]
adopted.head()
```

```python
#drop duplicate user ids
non_adopted = non_adopted.reset_index()
non_adopted = non_adopted.drop_duplicates(subset='user_id')
non_adopted.tail()
```

```python
adopted = adopted.reset_index()
adopted = adopted.drop_duplicates(subset='user_id')
adopted.head()
```

```python
#get hour from creation_time
hour = []
non_adopted = non_adopted.drop(columns=['index'])
non_adopted = non_adopted.reset_index()
non_adopted['creation_time'] = non_adopted['creation_time'].astype('str')
for j in range(len(non_adopted.creation_time)):
    hour.append(non_adopted['creation_time'][j][11:13])
    
```

```python
for k in range(len(hour)):
    hour[k] = int(hour[k])

non_adopted['hour'] = hour
non_adopted.head()
```

```python
plt.hist(non_adopted.hour, bins=24)
```
