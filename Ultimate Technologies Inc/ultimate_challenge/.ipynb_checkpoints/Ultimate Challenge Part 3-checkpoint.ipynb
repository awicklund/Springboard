To find the activity in the 6th month, we'll need to compute a column (active_time = last_trip_date - signup_date)

```python
#convert date columns to datetime datatype
df['last_trip_date'] = pd.to_datetime(df['last_trip_date'])
df['signup_date'] = pd.to_datetime(df['signup_date'])

#calculate active time
df['active_time'] = df.last_trip_date - df.signup_date
df.head(10)
```

```python
df.dtypes
```

While not exact, the approximate number of days in a month is 30. We want a bool column (active_status) to determine if a user is active at or after 180 days from signup.

```python
from datetime import datetime, timedelta
df['active_status'] = df.active_time >= timedelta(days=180)

#check if any users are active
df_active = df[df.active_status == True]
print(df_active)
```

Scale is still having difficulties processing the date columns. There may actually be colinearity at this point since start and active date are represented as active status. The two object columns will be removed.

```python
df = df.drop(columns = ['signup_date', 'last_trip_date'])
df.dtypes
```

```python
from sklearn import decomposition
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA

#Standardize the dataset
#https://etav.github.io/python/scikit_pca.html#:~:text=Principal%20Component%20Analysis%20%28PCA%29%20in%20Python%20using%20Scikit-Learn,amount%20of%20variance%20for%20a%20given%20data%20set.
x = df.values #convert the data into a numpy array
x = scale(x);x
```

```python
x.shape
```

```python
covar_matrix = PCA(n_components = 11) #we have 20 features
```

```python
covar_matrix.fit(x)
variance = covar_matrix.explained_variance_ratio_ #calculate variance ratios

var=np.cumsum(np.round(covar_matrix.explained_variance_ratio_, decimals=3)*100)
var #cumulative sum of variance explained with [n] features
```

```python
plt.ylabel('% Variance Explained')
plt.xlabel('# of Features')
plt.title('PCA Analysis')
plt.ylim(30,100.5)
plt.style.context('seaborn-whitegrid')


plt.plot(var)
```

PCA suggests 9 factors or more, so another method must be considered.

These two scatterplots show a general direction from left-bottom corner to right-upper corner. They are the most likely of the scatterplots to possibly fit a logistic curve based on shape. Weekday_pct has points all over the board for active users, so we'll create the logistic curve based on avg_rating_of_driver.

```python
#https://www.tutorialspoint.com/logistic_regression_in_python/logistic_regression_in_python_quick_guide.htm
#separate into X and Y 
X = df.iloc[:,:10]
X.head()
```

```python
Y = df.iloc[:,10]
Y.value_counts()
```

```python
#create test and training data
from sklearn import preprocessing
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=8)
```

```python
model = LogisticRegression(max_iter=400, random_state=8) #max_iter increased from 100 to 400 
#to resolve convergence issues with low retention
```

```python
model.fit(X_train, Y_train)
```

```python
pred_y = model.predict(X_test)
np.unique(pred_y)
```

```python
#find accuracy of logistic model
print('Accuracy: {:.2f}'.format(model.score(X_test, Y_test)))
```

The accuracy is too high for this model. The model predicts that a user will not be active in their 6th month ever (shown by pred_y's unique values), which makes sense given the very small value count for active users in Y. There is not enough data from active users past the 6th month mark to make a conclusion on what factor should be adjusted. Rather than a six month retention model, a model on a smaller interval (such as 4 month retention) may give clearer results.
